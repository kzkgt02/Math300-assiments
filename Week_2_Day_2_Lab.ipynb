###  Problem 5
Let

$
A=
\begin{bmatrix}
2 & 1 \\
3 & 4
\end{bmatrix},
\qquad
B=
\begin{bmatrix}
5 & 2 \\
1 & 6
\end{bmatrix}
$

Compute the product.

````python
A = np.array([[2, 1],
              [3, 4]])

B = np.array([[5, 2],
              [1, 6]])

matrix_product = A @ B

print("Matrix Product (A @ B):\n", matrix_product)
````
