### Problem 2
1. Compute the correlation between all the features.
2. What is the largest correlated/anti-correlated features?
3. Do you have an explanation for why?

```python
corr = df.corr(numeric_only=True)
print(corr)
corr_pairs = corr.unstack()

corr_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]

print("\nLargest Positive Correlation:")
print(corr_pairs.sort_values(ascending=False).head())

print("\nLargest Negative Correlation:")
print(corr_pairs.sort_values().head())
```

### Problem 3
Create one dataset with a strong positive correlation and another with a strong negative correlation. Plot them side by side and compare their correlation coefficients.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

x1 = np.arange(100)
y1 = x1 + np.random.normal(0,8,100)

x2 = np.arange(100)
y2 = -x2 + np.random.normal(0,8,100)

corr1 = np.corrcoef(x1,y1)[0,1]
corr2 = np.corrcoef(x2,y2)[0,1]

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(x1,y1)
plt.title(f"Positive Correlation\nr={corr1:.2f}")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(1,2,2)
plt.scatter(x2,y2)
plt.title(f"Negative Correlation\nr={corr2:.2f}")
plt.xlabel("X")
plt.ylabel("Y")

plt.tight_layout()
plt.show()

print("Positive correlation:", corr1)
print("Negative correlation:", corr2)
```
