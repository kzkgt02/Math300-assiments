## Problem 2

1. Write a recursive function `factorial()` that takes in a natural number `n` and computes $n!=n(n-1)(n-2)\cdots3\cdot2\cdot 1$.
2. Use `%timeit` to determine how fast your function calculates `factorial(150)`.
3. Use `%timeit` to calculate how fast `math.factorial()` calculates the same quantity.
4. Were there any differences between the two methods?

```python
%timeit math.factorial(150)
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
%timeit factorial(4)
```

```python
%timeit factorial(150)
```

```python
%timeit math.factorial(150)
```

```python
#math.factorial() is much faster because it is implemented in optimized C code, while the recursive function repeatedly makes Python function calls, which adds overhead.
```

## Problem 3
Recall for a quadratic equation of the form
$$ax^2 + bx + c = 0
$$

The solutions are given by the <i>quadratic formula</i>:

$$
x=\frac{-b\pm\sqrt{b^2-4ac}}{2a}
$$

1. Write a function `quadratic()` that calculates the two roots given by the quadratic formula only if they are real.
2. Use `%timeit` to determine how fast your function calculates the roots for $437x^2 - 915x + 478 = 0$.
3. Use `%timeit` to calculate how fast `sp.solve()` calculates the same quantities.
4. Were there any differences between the two methods?

```python
def quadratic(a, b, c):
    d = b**2 - 4*a*c

    if d < 0:
        return "No real roots"

    x1 = (-b + math.sqrt(d)) / (2*a)
    x2 = (-b - math.sqrt(d)) / (2*a)

    return x1, x2
```

```python
quadratic(437, -915, 478)
```

```python
%timeit quadratic(437, -915, 478)
```

```python
x = sp.symbols('x')

%timeit sp.solve(437*x**2 - 915*x + 478, x)
```

```python
#The custom function is much faster because it directly applies the quadratic formula. sympy.solve() performs symbolic computation, making it more flexible but slower.
```
