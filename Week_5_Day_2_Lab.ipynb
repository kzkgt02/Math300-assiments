### Problem 1
Suppose $f(x)=x^2+6x+10$

1. Plot the function to visualize the minimum
2. Find the minimum using `scipy`

````python
#1
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

x = np.linspace(-10,4,300)
y = x**2 + 6*x + 10

plt.figure(figsize=(6,5))

plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x)=x²+6x+10")

plt.grid()

plt.show()
````

````python
#2
def f(x):
    return x**2 + 6*x + 10

result = minimize(f, x0=0)

print("Minimum x =", result.x)
print("Minimum value =", result.fun)
````


## Problem 2

Find the maximum of $50x-x^2$.

````python
from scipy.optimize import minimize

def f(x):
    return -(50*x - x**2)

result = minimize(f, x0=0)

print("Maximum x =", result.x)
print("Maximum value =", -result.fun)
````

## Problem 3

Minimize

```python
np.sin(x)+x**2
```

using different starting values.

Do you obtain the same solution? Compute the derivative using `sympy`. When is this derivative zero?

````python
from scipy.optimize import minimize

def f(x):
    return np.sin(x) + x**2

starts = [-3,-1,0,2,5]

for s in starts:
    result = minimize(f,x0=s)
    print("Start:",s)
    print("Minimum x =",result.x)
    print("Minimum value =",result.fun)
    print()
````

````python
import sympy as sp

x = sp.Symbol('x')

f = sp.sin(x) + x**2

derivative = sp.diff(f,x)

print(derivative)
````

````python
solution = sp.nsolve(derivative,0)

print(solution)
````

````python
# Different starting values give the same minimum because this function has one global minimum. The derivative is 2x+cos(x), and it is zero at approximately x = -0.4502.
````

### Problem 4
Design a function with multiple minima and optimize.

````python
# multiple minima
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def f(x):
    return np.sin(3*x) + x**2/5

x = np.linspace(-5,5,400)

plt.figure(figsize=(6,5))

plt.plot(x,f(x))

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Function with Multiple Minima")

plt.grid()

plt.show()
````

````python
# optimize
starts = [-4,-2,0,2,4]

for s in starts:
    result = minimize(f,x0=s)
    print("Start:",s)
    print("Minimum x =",result.x)
    print("Minimum value =",result.fun)
    print()
````
