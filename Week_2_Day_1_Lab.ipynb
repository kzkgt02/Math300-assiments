### Problem 5
Write a recursive function `fibonacci(n)` that computes the $n$-th Fibonacci number. Compute the time complexity of your `fibonacci(n)` function.

````python
def fibonacci(n):
    if n < 1:
        return 'Need a positive number'
    elif n == 1 or n == 2:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)
````

````python
x = 1
while x < 10:
    print(fibonacci(x))
    x += 1
````

### Problem 6
Write a recursive function `reverse(string)` that reverses the order of a string. Compute the time complexity of your `reverse()` function

````python
def reverse(string):
    if string == '':
        return string
    else:
        return reverse(string[1:])+string[0]
````

### Problem 7
Write a recursive function that computes the sum of integers up to input $n$.

````python
def partial_sum(n):
    if n < 0:
        return print('Need positive integer')
    elif n == 0:
        return 0
    else:
        return partial_sum(n-1) + n
````

````python
x = 2
while x < 10:
    print(partial_sum(x))
    x += 1
````
