### Problem 1
Write your own function performs vectors addition on lists instead of arrays.

```python
def vector_add(v1,v2):
    result = []
    for i in range(len(v1)):
        result.append(v1[i] + v2[i])
    return result

a = [1,2,3]
b = [4,5,6]
vector_add(a,b)
```

### Problem 3
Make a 2 by 2 invertible matrix, A.
1. Compute its eigenvectors and plot them using matplotlib
2. Compute A times the eigenvectors and plot them on the same plot.
3. What do you notice?

```python
#1
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[2,1],
              [1,2]])

eigenvalues, eigenvectors = np.linalg.eig(A)

v1 = eigenvectors[:,0]
v2 = eigenvectors[:,1]

plt.figure(figsize=(6,6))

plt.quiver(0,0,
           v1[0],v1[1],
           angles='xy',
           scale_units='xy',
           scale=1,
           color='blue',
           label='Eigenvector 1')

plt.quiver(0,0,
           v2[0],v2[1],
           angles='xy',
           scale_units='xy',
           scale=1,
           color='red',
           label='Eigenvector 2')

plt.xlim(-2,2)
plt.ylim(-2,2)

plt.grid()
plt.legend()

plt.title("Eigenvectors")

plt.show()
```

```python
#2
Av1 = A @ v1
Av2 = A @ v2

plt.figure(figsize=(6,6))

plt.quiver(0,0,
           v1[0],v1[1],
           angles='xy',
           scale_units='xy',
           scale=1,
           color='blue',
           label='Eigenvector 1')

plt.quiver(0,0,
           v2[0],v2[1],
           angles='xy',
           scale_units='xy',
           scale=1,
           color='red',
           label='Eigenvector 2')

plt.quiver(0,0,
           Av1[0],Av1[1],
           angles='xy',
           scale_units='xy',
           scale=1,
           color='cyan',
           label='A*v1')

plt.quiver(0,0,
           Av2[0],Av2[1],
           angles='xy',
           scale_units='xy',
           scale=1,
           color='orange',
           label='A*v2')

plt.xlim(-4,4)
plt.ylim(-4,4)

plt.grid()
plt.legend()

plt.title("Eigenvectors and A*Eigenvectors")

plt.show()
```

```python
#3 The vectors A·v1 and A·v2 have the same direction as the original eigenvectors. Only their lengths change because the eigenvectors are multiplied by their eigenvalues.
```
