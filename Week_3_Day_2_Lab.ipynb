### Problem 1
Plot the function $f(x)=x^2$ for $x\in [-1,1]$ using `np.linspace()` with 10, 100, 1000 slices. Make sure your plot is as informative as possible.

```python
import numpy as np
import matplotlib.pyplot as plt

x10 = np.linspace(-1, 1, 10)
y10 = x10**2

x100 = np.linspace(-1, 1, 100)
y100 = x100**2

x1000 = np.linspace(-1, 1, 1000)
y1000 = x1000**2

plt.plot(x10, y10, label="10 slices")
plt.plot(x100, y100, label="100 slices")
plt.plot(x1000, y1000, label="1000 slices")

plt.title("Plot of $f(x)=x^2$")
plt.xlabel("x")
plt.ylabel("x²")
plt.legend()

plt.show()
```
